```python
# =========================================
# SIMULADOR IA DENTISTA
# Menú interactivo para preparatoria
# Uso de: if, elif, else y while
# =========================================

print("===================================")
print("     IA DENTISTA - CONSULTORIO")
print("===================================")

while True:

    print("\nMENU PRINCIPAL")
    print("1. Revisar dolor dental")
    print("2. Revisar sensibilidad")
    print("3. Revisar sangrado")
    print("4. Ver consejo dental")
    print("5. Salir")

    opcion = input("Selecciona una opcion: ")

    # OPCION 1
    if opcion == "1":

        print("\n--- REVISION DE DOLOR ---")

        dolor = input("¿Tienes dolor fuerte? (si/no): ")

        if dolor == "si":
            print("IA Dentista: Posible caries o infeccion.")
            print("Recomendacion: Visita al dentista pronto.")
        else:
            print("IA Dentista: Tu dolor parece leve.")
            print("Recomendacion: Cepillate correctamente.")

    # OPCION 2
    elif opcion == "2":

        print("\n--- REVISION DE SENSIBILIDAD ---")

        frio = input("¿Te duelen los dientes con cosas frias? (si/no): ")

        if frio == "si":
            print("IA Dentista: Posible sensibilidad dental.")
            print("Usa pasta dental para dientes sensibles.")
        else:
            print("IA Dentista: No se detecta sensibilidad.")

    # OPCION 3
    elif opcion == "3":

        print("\n--- REVISION DE SANGRADO ---")

        sangre = input("¿Tus encias sangran? (si/no): ")

        if sangre == "si":
            print("IA Dentista: Posible gingivitis.")
            print("Debes mejorar tu higiene bucal.")
        else:
            print("IA Dentista: Tus encias parecen saludables.")

    # OPCION 4
    elif opcion == "4":

        print("\n--- CONSEJO DENTAL ---")

        print("✓ Cepilla tus dientes 3 veces al dia.")
        print("✓ Usa hilo dental.")
        print("✓ Evita mucho azucar.")
        print("✓ Visita al dentista cada 6 meses.")

    # OPCION 5
    elif opcion == "5":

        print("\nGracias por usar IA Dentista.")
        print("Programa finalizado.")
        break

    # OPCION INCORRECTA
    else:

        print("\nOpcion no valida.")
        print("Intenta nuevamente.")
```
